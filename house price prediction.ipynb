Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
```

Importing the Botson House Price Dataset

```python
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# load the dataset to a pandas
house_price_dataframe=pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
#print first five rows of our dataframe
house_price_dataframe.head()
```

```python
# add the target coulum (price)to dataframe
house_price_dataframe['price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking the number of rows and columns in the data frame
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of datset
house_price_dataframe.describe()
```

Understanding correlation between the data


```python
#postive correaltion and negative correlation
correlation=house_price_dataframe.corr()
```

```python
#constructing a heap map for understanding  the correaltion
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Splitting the data and the target

```python
X=house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training
XGboost Regressor

```python
#loading the model
model=XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

Prediction on Training Data

```python
#accuracy for prediction on training data
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R sqaure ERROR
score_1=metrics.r2_score(Y_train,training_data_prediction)

#mean absolute error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print('R score:',score_1)
print('mean absolute error:',score_2)
```

Visualising the actual Prices with Predicted Vlaues

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("actual price v/s predicted price")
plt.show()
```

Prediction on Test Data

```python
#accuracy for prediction on test data
test_data_prediction=model.predict(X_test)
```

```python
# R sqaure ERROR
score_1=metrics.r2_score(Y_test,test_data_prediction)

#mean absolute error
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)
print('R score:',score_1)
print('mean absolute error:',score_2)
```
